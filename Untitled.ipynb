```python
def generalized_dice_loss(im_real, im_pred):
    eps = 1e-9
    total_voxels = im_real.shape[0] * im_real.shape[1] * im_real.shape[2]
    
    print(f'total:\t{total_voxels}\t')
    [print(np.sum(np.where(im_real == index, 1, 0)) / total_voxels) for index in range(3)]
    print()
    
    weights = [1 / np.power(np.sum(np.where(im_real == index, 1, 0)), 2) for index in range(3)]
    
    
    print(weights)
    
    numerator = [np.sum(np.where(im_real==index, 1, 0) * np.where(im_pred==index, 1, 0)) * weights[index] for index in range(3)]
    print(numerator)
    
    denominator = [np.sum(np.where(im_real==index, 1, 0) + np.where(im_pred==index, 1, 0)) + weights[index] for index in range(3)]
    print(denominator)
    
    return 1 - ((2 * sum(numerator)) / (sum(denominator) + eps))
    
# a = np.random.randn(1, 512, 512, 128)
a = nib.load(data_meta['training'][0]['label']).get_fdata()[:, :, :20]
b = nib.load(data_meta['training'][2]['label']).get_fdata()[:, :, :20]
# b = np.ones_like(a) +1 
# b = a.copy() + 2

generalized_dice_loss(a, b)
```
